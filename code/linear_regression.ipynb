```python
import pandas as pd
```

```python
# 데이터 불러오기
score = pd.read_csv('../data/삼성전자_감성점수_일별.csv', parse_dates=['날짜'])
samsung = pd.read_csv('../data/삼성전자_일별.csv', parse_dates=['Date'])
kospi = pd.read_csv('../data/코스피_일별.csv', parse_dates=['Date'])
kospi_comm = pd.read_csv('../data/코스피200_커뮤니케이션_일별.csv', parse_dates=['날짜'])
kospi_const = pd.read_csv('../data/코스피200_건설_일별.csv', parse_dates=['날짜'])
kospi_heav = pd.read_csv('../data/코스피200_중공업_일별.csv', parse_dates=['날짜'])
kospi_stee = pd.read_csv('../data/코스피200_철강소재_일별.csv', parse_dates=['날짜'])
kospi_ener = pd.read_csv('../data/코스피200_에너지화학_일별.csv', parse_dates=['날짜'])
kospi_info = pd.read_csv('../data/코스피200_정보기술_일별.csv', parse_dates=['날짜'])
kospi_fina = pd.read_csv('../data/코스피200_금융_일별.csv', parse_dates=['날짜'])
kospi_cons = pd.read_csv('../data/코스피200_생활소비재_일별.csv', parse_dates=['날짜'])
kospi_disc = pd.read_csv('../data/코스피200_경기소비재_일별.csv', parse_dates=['날짜'])
kospi_indu = pd.read_csv('../data/코스피200_산업재_일별.csv', parse_dates=['날짜'])
kospi_heal = pd.read_csv('../data/코스피200_헬스케어_일별.csv', parse_dates=['날짜'])
```

```python
# 데이터 전처리(날짜 맞춤)
while len(score) > len(samsung):
    for i in range(len(samsung)):
        if score.iloc[i, 0] != samsung.iloc[i, 0]:
            score = score.drop(i, axis=0).reset_index(drop=True)
            break

score = score.drop(221, axis=0).reset_index(drop=True)
samsung = samsung.drop(0, axis=0).reset_index(drop=True)
kospi = kospi.drop(0, axis=0).reset_index(drop=True)

# 2022년 5월 9일 데이터 삭제. krx 데이터에는 해당 날짜에 대한 데이터가 있으나, 야후 파이낸스 데이터에는 해당 날짜에 대한 데이터가 없음.
kospi_comm = kospi_comm.drop(81, axis=0).reset_index(drop=True)
kospi_const = kospi_const.drop(81, axis=0).reset_index(drop=True) 
kospi_heav = kospi_heav.drop(81, axis=0).reset_index(drop=True) 
kospi_stee = kospi_stee.drop(81, axis=0).reset_index(drop=True) 
kospi_ener = kospi_ener.drop(81, axis=0).reset_index(drop=True)
kospi_info = kospi_info.drop(81, axis=0).reset_index(drop=True)
kospi_fina = kospi_fina.drop(81, axis=0).reset_index(drop=True) 
kospi_cons = kospi_cons.drop(81, axis=0).reset_index(drop=True) 
kospi_disc = kospi_disc.drop(81, axis=0).reset_index(drop=True) 
kospi_indu = kospi_indu.drop(81, axis=0).reset_index(drop=True)
kospi_heal = kospi_heal.drop(81, axis=0).reset_index(drop=True)

score['날짜'] = samsung['Date']
```

```python
# 전처리 결과 검토
for i in range(len(samsung)):
    if score['날짜'][i] != kospi['Date'][i]:
        print('Error')
    if score['날짜'][i] != kospi_comm['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_const['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_heav['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_stee['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_ener['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_info['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_fina['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_cons['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_disc['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_indu['날짜'][i]:
        print('Error')
    if score['날짜'][i] != kospi_heal['날짜'][i]:
        print('Error')
```

```python
import copy

df = copy.deepcopy(score)
```

```python
df['삼성전자'] = samsung['Adj Close'].to_list()
df['KOSPI'] = kospi['Adj Close'].to_list()
df['커뮤니케이션서비스'] = kospi_comm['종가'].to_list()
df['건설'] = kospi_const['종가'].to_list()
df['중공업'] = kospi_heav['종가'].to_list()
df['철강/소재'] = kospi_stee['종가'].to_list()
df['에너지/화학'] = kospi_ener['종가'].to_list()
df['정보기술'] = kospi_info['종가'].to_list()
df['금융'] = kospi_fina['종가'].to_list()
df['생활소비재'] = kospi_cons['종가'].to_list()
df['경기소비재'] = kospi_disc['종가'].to_list()
df['산업재'] = kospi_indu['종가'].to_list()
df['헬스케어'] = kospi_heal['종가'].to_list()
```

```python
df = df.set_index(keys='날짜', drop=True)
```

```python
df
```

```python
import statsmodels.api as sm
```

```python
# 1번 모형, y: 삼성전자, x: KOSPI200 전체 산업군 지수, 삼성전자 감성점수
Y = df['삼성전자']
df_X = df.drop(['삼성전자', 'KOSPI'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
# Backward Elimination
df_X = df.drop(['삼성전자', 'KOSPI', '생활소비재', '철강/소재', '중공업', '커뮤니케이션서비스', '헬스케어', '건설', '금융', '경기소비재'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
# 최적 모형에서 감성 점수를 제외한 1번 모형
df_X = df.drop(['삼성전자', 'KOSPI', '생활소비재', '철강/소재', '중공업', '커뮤니케이션서비스', '헬스케어', '건설', '금융', '경기소비재', '감성점수'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
# 2번 모형, y: KOSPI, x: KOSPI200 전체 산업군 지수, 감성점수
Y = df['KOSPI']
df_X = df.drop(['삼성전자', 'KOSPI'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
# Backward Elimination
df_X = df.drop(['삼성전자', 'KOSPI', '건설', '생활소비재'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

```python
# 최적 모형에서 감성 점수를 제외한 2번 모형
df_X = df.drop(['삼성전자', 'KOSPI', '건설', '생활소비재', '감성점수'], axis=1)
X = sm.add_constant(df_X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```
